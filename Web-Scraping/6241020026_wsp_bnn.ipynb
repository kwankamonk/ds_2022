```python
product=s.find_all('a',class_='product-item')
brands=[]
names=[]
debs=[]
prices=[]
discounts=[]



for f in product:
    brand=f.find('div',class_='product-label-brand')
    brands.append(brand)
    name= f.find('div',class_='product-name')
    names.append(name)
    deb=f.find('div',class_='product-short-attribute') if f.find('div',class_='product-short-attribute') else None 
    debs.append(deb)
    price=f.find('div',class_='product-price')
    prices.append(price)
    discount=f.find('strong') if f.find('strong') else None 
    discounts.append(discount)



df= pd.DataFrame({
    'brand':brands,
    'name':names,
    'deb':debs,
    'price':prices,
    'discount':discounts,


})
df[:5]
```

```python
product=s.find_all('a',class_='product-item')
brands=[]
names=[]
debs=[]
prices=[]
discounts=[]
imgs=[]
links=[]


for f in product:
    brand=f.find('div',class_='product-label-brand').text.strip()
    brands.append(brand)
    name= f.find('div',class_='product-name').text.strip()
    names.append(name)
    deb=f.find('div',class_='product-short-attribute').text.strip() if f.find('div',class_='product-short-attribute') else None
    debs.append(deb)
    price=f.find('div',class_='product-price').text.strip()
    prices.append(price)
    discount=f.find('strong').text.strip() if f.find('strong') else None
    discounts.append(discount)
    img=f.find('img',class_='image')['src']
    imgs.append(img)
    link=f"https://www.bnn.in.th{f['href']}"
    links.append(link)


df= pd.DataFrame({
    'brand':brands,
    'name':names,
    'deb':debs,
    'price':prices,
    'discount':discounts,
    'url_img':imgs,
    'url_link':links

})
df[:5]
```

#### แสดงรูปที่อยู่ใน dataframe

```python
from IPython.display import HTML
```

```python
def to_html_img(url, alt=None, width=100): #เพื่อคลุม image url #width เพื่อรูปไม่ใหญ่เกิน
    if alt==None:
        alt=''
    return f'<img alt="{alt}" src="{url}" width="{width}"/>'
```

```python
to_html_img('https://media-cdn.bnn.in.th/170213/Microsoft-Surface-Pro8-i5-Thai-Platinum-(8PN-00016)-1-square_medium.jpg')
```

```python
df['img']=df.apply(lambda r: to_html_img(r['url_img'],r['name'], 100), axis=1)
df[:2]
```

```python
HTML(df[:2][['name','img']].to_html(escape=False))
```

#### กดรูปแล้วเข้าปลายทาง

```python
def to_html_l(url, text): #กดรูป แล้วไปหน้านั้นๆ
    if text==None:
        text=''
    return f'<a href="{url}">{text}</a>'
```

```python
to_html_l('https://www.bnn.in.th//th/p/smartphone-and-accessories/smartphone/samsung-smartphone/samsung-smartphone-galaxy-a52s-8128gb-awesome-black-n-5g-8806092968752_z78y0w', '<img alt="สมาร์ทโฟน Samsung Galaxy A52s (8+128GB) Awesome Black-N (5G)" src="https://media-cdn.bnn.in.th/151798/Samsung-Smartphone-Galaxy-A52s-5G-1-square_medium.jpg" width="200"/>')
```

```python
df['link']=df.apply(lambda r: to_html_l(r['url_link'], r['name']),axis=1)
df[:2]
```
